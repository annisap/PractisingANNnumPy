```python
"""To use BFGS, the minimize function requires us to pass in an objective function
"""
%pylab inline
#Import code from previous videos:
from partFive import *

from scipy import optimize
```

```python
"""The objective function is the costFunctionWrapper. 
It accepts a vector of parameters, input data, and output data, and returns both the cost and gradients
"""

class trainer(object):
    def __init__(self, N):
        #Local Reference to the network
        self.N = N
        
    #callback tracks the cost function value as we train the network
    def callbackF(self, params):
        self.N.setParams(params)
        self.J.append(self.N.costFunction(self.X, self.y))

    def costFunctionWrapper(self, params, X, y):
        self.N.setParams(params)
        cost = self.N.costFunction(X, y)
        grad = self.N.computeGradients(X, y)
        return cost, grad
        
    def train(self, X, y):
        #Make internal variable for the callback function
        self.X = X
        self.y = y
        
        # Make an empty list to store the costs:
        self.J = []
        
        params0 = self.N.getParams()
        
        options = {'maxiter': 200, 'disp': True}
        
        #set the jacobian parameter to true since weâ€™re computing the gradient within our neural network class
        _res = optimize.minimize(self.costFunctionWrapper, params0, jac=True, method='BFGS', \
                                 args=(X, y), options=options, callback=self.callbackF)
        
        #Once the network is trained, replace the original, random parameters, with the trained parameters.
        self.N.setParams(_res.x)
        self.optimizationResults = _res
```
